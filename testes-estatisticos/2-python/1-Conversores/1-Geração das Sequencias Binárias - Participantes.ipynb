```python
import pandas as pd
import numpy as np
import warnings

pd.options.display.float_format = '{:.4f}'.format
warnings.filterwarnings("ignore")

file_path = "2-resultados_sorteios.csv"
output_path = "3-seqbinaria_participantes.csv"

try:
    df = pd.read_csv(file_path, sep="\t", engine="python")

    # Renomeia as colunas
    colunas = ["Inscrição", "Nome"] + [f"Sorteio {i}" for i in range(1, df.shape[1] - 1)]
    df.columns = colunas

    # Converter as colunas dos sorteios para numérico
    df.iloc[:, 2:] = df.iloc[:, 2:].apply(pd.to_numeric, errors='coerce')

    with open(output_path, "w") as f:
        for idx, row in df.iterrows():
            dados_sem_na = row[2:].dropna()
            mediana = np.median(dados_sem_na)
            seq_binaria = np.where(dados_sem_na > mediana, 1, 0)
            seq_str = ''.join(seq_binaria.astype(str))
            linha = f"Candidato {idx + 1}\t{mediana}\t{seq_str}"
            f.write(linha + "\n")

    print(f"Arquivo '{output_path}' gerado com sucesso.")

except FileNotFoundError:
    print("Arquivo não encontrado.")
```
