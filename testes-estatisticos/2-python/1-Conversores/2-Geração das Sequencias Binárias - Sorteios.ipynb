```python
import pandas as pd
import numpy as np
import warnings

pd.options.display.float_format = '{:.4f}'.format
warnings.filterwarnings("ignore")

file_path = "2-resultados_sorteios.csv"
output_path = "4-seqbinaria_sorteios.csv"

try:
    df = pd.read_csv(file_path, sep="\t", engine="python")

    # Renomeia as colunas (assumindo 2 primeiras: Inscrição e Nome, o resto são sorteios)
    colunas = ["Inscrição", "Nome"] + [f"Sorteio {i}" for i in range(1, df.shape[1] - 1)]
    df.columns = colunas

    with open(output_path, "w") as f:
        # Para cada sorteio (coluna)
        for sorteio in colunas[2:]:
            dados = df[sorteio].dropna()
            mediana = np.median(dados)
            seq_binaria = np.where(dados > mediana, 1, 0)
            seq_str = ''.join(seq_binaria.astype(str))
            linha = f"{sorteio}\t{mediana}\t{seq_str}"
            f.write(linha + "\n")

    print(f"Arquivo '{output_path}' gerado com sucesso.")

except FileNotFoundError:
    print("Arquivo não encontrado.")
```
