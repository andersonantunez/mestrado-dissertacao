```python
import numpy as np
from scipy.stats import entropy

# Função usando scipy.stats.entropy
def teste_entropia(seq_str):
    values, counts = np.unique(list(seq_str), return_counts=True)
    probabilities = counts / counts.sum()
    return entropy(probabilities, base=2)


# Primeiro exemplo do livro, página 60
print(f"--------------------------------------------- Exemplo 1")
seq_str = "000100010000100000100100"
ent = teste_entropia(seq_str)
print(f"Sequência: {seq_str}")
print(f"Tamanho: {len(seq_str)}")
print(f"Entropia de Shannon: {ent:.4f} bits")
print()

# Segundo exemplo do livro, página 61
print(f"--------------------------------------------- Exemplo 2")
seq_str = "111100100010111010110010"
ent = teste_entropia(seq_str)
print(f"Sequência: {seq_str}")
print(f"Tamanho: {len(seq_str)}")
print(f"Entropia de Shannon: {ent:.4f} bits")
print()

# Terceiro exemplo do livro, página 61
print(f"--------------------------------------------- Exemplo 3")
seq_str = "1110110001101111111011110"
ent = teste_entropia(seq_str)
print(f"Sequência: {seq_str}")
print(f"Tamanho: {len(seq_str)}")
print(f"Entropia de Shannon: {ent:.4f} bits")
print()
```
